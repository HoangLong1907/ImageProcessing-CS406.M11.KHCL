```python
def threshold(img_src, T):
  img_rst = img_src.copy()
  w, h, c = img_src.shape
  for i in range (w):
    for j in range (h):
      for k in range (c):
        if img_src[i][j][k] < T:
          img_rst[i][j][k] = 255
        else:
          img_rst[i][j] = 0
  return img_rst
```

```python
import matplotlib.pyplot as plt
def show2image(img_src, img_rst, title1, title2):
  plt.figure(figsize = (20, 20))
  plt.subplot(1, 2, 1)
  plt.title(title1)
  img_src = cv.cvtColor(img_src, cv.COLOR_BGR2RGB)
  img_src = plt.imshow(img_src, interpolation = 'bicubic')

  plt.subplot(1, 2, 2)
  plt.title(title2)
  img_rst = cv.cvtColor(img_rst, cv.COLOR_BGR2RGB)
  img_rst = plt.imshow(img_rst, interpolation = 'bicubic')
```

```python
img2 = cv2.imread("/content/drive/MyDrive/CS406.M11.KHCL/Baitap/Baitap1/images/pout.tif")
cv2_imshow(img2)
```

```python
def Bright(img, contrast, brightness):
  img_copy = img.copy()
  w, h, d = img.shape
  for k in range(d):
    for j in range(h):
      for i in range(w):
        img_copy[i][j][k] = (img[i][j][k] - 0.5) * contrast + 0.5 + brightness 
        if img_copy[i][j][k] > 255: img_copy[i][j][k] = 255
        if img_copy[i][j][k] < 0: img_copy[i][j][k] = 0
  return img_copy
```

```python
show2image(img2,Bright(img2, 0.5 100), "Before", "After")
```

```python
img4 = cv2.imread("/content/drive/MyDrive/CS406.M11.KHCL/Baitap/Baitap1/images/kids.tif")
cv2_imshow(img4)
```

```python
import numpy as np

def invert(f, L = 256): #L: gi tr mc xm ln nht
# cng thc : g(x,y) = L-1-f(x,y)
    new_img = np.zeros(f.shape, f.dtype)
    for i in range(f.shape[0]):
        for j in range(f.shape[1]):
            for k in range(f.shape[2]):
                new_img[i,j,k] = L-1-f[i,j,k]
    return new_img
```

```python
show2image(img4, invert(img4), "Before", "After")
```
