```python
import matplotlib.pyplot as plt
import cv2

def show2image(img_src, img_rst, title1, title2):
  plt.figure(figsize = (20, 20))
  plt.subplot(1, 2, 1)
  plt.title(title1)
  img_src = cv2.cvtColor(img_src, cv2.COLOR_BGR2RGB)
  img_src = plt.imshow(img_src, interpolation = 'bicubic')

  plt.subplot(1, 2, 2)
  plt.title(title2)
  img_rst = cv2.cvtColor(img_rst, cv2.COLOR_BGR2RGB)
  img_rst = plt.imshow(img_rst, interpolation = 'bicubic')
```

```python
from google.colab.patches import cv2_imshow
img1 = cv2.imread(path_img3,1)
cv2_imshow(img1)
```

# Erosion and Dilation

```python
import numpy as np

kernel = np.ones((5,5), np.uint8)

img_erosion = cv2.erode(img1, kernel, iterations=1)

img_dilation = cv2.dilate(img1, kernel, iterations=1)
```

```python
show2image(img1, img_erosion, 'Before', 'After')
```

```python
show2image(img1, img_dilation, 'Before', 'After')
```

# Opening

```python
opening = cv2.morphologyEx(img1, cv2.MORPH_OPEN, kernel)
```

```python
show2image(img1, opening, 'Before', 'After')
```

# Closing

```python
closing = cv2.morphologyEx(img1, cv2.MORPH_CLOSE, kernel)
```

```python
show2image(img1, closing, 'Before', 'After')
```

# Get Coins

```python
img4 = cv2.imread(path_img4)
cv2_imshow(img4)
```

```python
T, thresh = cv2.threshold(img4, 50, 255, cv2.THRESH_BINARY)
kernel = np.ones((2,2),np.uint8)
img4_opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
img4_dilation = cv2.dilate(img4_opening,kernel, iterations=1)
show2image(img4,img4_dilation, 'Before', 'After' )
```

# Vân tay

```python
img1 = cv2.imread(path_img)
cv2_imshow(img1)
```

```python
(T, thresh) = cv2.threshold(img1, 70, 255, cv2.THRESH_BINARY)
kernel = np.ones((3,3), np.uint8)
img1_erosion = cv2.erode(thresh, kernel, iterations=1)
img1_dilation = cv2.dilate(img1_erosion, kernel, iterations=2)
img1_result = cv2.erode(img1_dilation, kernel, iterations=1)
show2image(img1, img1_result, 'Before', 'After')
```
